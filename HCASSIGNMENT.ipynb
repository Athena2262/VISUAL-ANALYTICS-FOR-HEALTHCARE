```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
df = pd.read_csv('/content/sleep.csv')
df.head()
```

```python
df.info()
```

```python
mean = df['Sleep Duration'].mean()
var = df['Sleep Duration'].var()
std = df['Sleep Duration'].std()
print (f"Mean {mean}\nVariance {var}\nStandard Deviation = {std}")
```

```python
mean = df['Physical Activity Level'].mean()
var = df['Physical Activity Level'].var()
std = df['Physical Activity Level'].std()
print (f"Mean {mean}\nVariance {var}\nStandard Deviation = {std}")
```

```python
plt.figure(figsize=(10,2))
plt.boxplot(df['Stress Level'], vert=False, showmeans=True)
plt.grid(color='blue', linestyle='dotted')
plt.show()
```

```python
sns.lineplot(x='Stress Level', y='Sleep Duration', data=df)
plt.show()
```

```python
sns.scatterplot(x='Sleep Duration', y='Physical Activity Level', data=df,
               hue='Heart Rate')
plt.show()
```

```python
plt.figure(figsize=(10,2))
plt.boxplot(df['Heart Rate'], vert=False, showmeans=True)
plt.grid(color='blue', linestyle='dotted')
plt.show()
```

```python
fig = px.scatter(df, x='Stress Level', y='Blood Pressure', title='Scatter Plot')
fig.show()
```

```python
sns.barplot(x='Sleep Duration', y='Physical Activity Level', data=df,
               hue='Heart Rate')

plt.show()
```
